```python
from PIL import Image
import os

workingDir = './xxx/xxx/'
savingDir = './xxx/xxx/xxx/xxx/xxx/'
names = [name for name in os.listdir("./xxx/xxx/")]

def Brightness(imgname):
    fullPath = workingDir+imgname
    savePath = savingDir+imgname
    im = Image.open(fullPath)
    pix = im.load()
    I, J = im.size
    brightness = []
    for i in range(I):
        for j in range(J):
            if pix[i,j][1] > 5:
                brightness.append(pix[i,j][1])
    avgBrightness = sum(brightness)/len(brightness)
    if avgBrightness < 70:
        multiplier = 70/avgBrightness
    else:
        multiplier = 1
    for i in range(I):
        for j in range(J):
            tmp = list(pix[i,j])
            if  tmp[1] > 5:
                tmp[1] = min(255, int(tmp[1]*multiplier))
                tmp[0] = 0
            pix[i, j] = tuple(tmp)
    print ("OK!")
    im.save(savePath[:-4]+'_new.jpg') 
for nm in names:
    Brightness(nm)
```
